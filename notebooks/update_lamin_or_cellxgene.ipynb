```python
import lamindb as ln

files = ln.Collection.using(instance="laminlabs/cellxgene").filter(name="cellxgene-census").first().artifacts.all()
for i, f in enumerate(files[1:]):
    print(i, f)
    if len(ln.Artifact.filter(uid=f.uid))==1:
        print("already cached")
        continue
    if not f.is_latest:
        print("not latest version")
        continue
    backed = f.open()
    if backed.obs.is_primary_data.sum() == 0:
        print("only contains non primary cells.. dropping")
        continue
    f.cache()
    if len(ln.Artifact.filter(uid=f.uid))==0:
        try:
            f.save()
        except Exception as e:
            if len(ln.Artifact.filter(uid=f.uid))==0:
                raise e
```
